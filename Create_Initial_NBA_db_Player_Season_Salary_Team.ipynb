## Create Tables

```python
import sqlite3
conn = sqlite3.connect('NBA_Fantasy_db.sqlite')
cur = conn.cursor()

cur.executescript('''
DROP TABLE IF EXISTS Salary;
DROP TABLE IF EXISTS Season;
DROP TABLE IF EXISTS Player;
DROP TABLE IF EXISTS Team
''')

cur.executescript('''
CREATE TABLE Player (
    id INTEGER NOT NULL PRIMARY KEY UNIQUE,
    name VARCHAR(40),
    name_ascii VARCHAR(40),
    code VARCHAR(20) UNIQUE
);
CREATE TABLE Team (
    id INTEGER NOT NULL PRIMARY KEY UNIQUE,
    name VARCHAR(32),
    code CHAR(3)
);
CREATE TABLE Salary (
    player_id INTEGER,
    season_id INTEGER,
    amount INTEGER,
    PRIMARY KEY (player_id, season_id) 
);
CREATE TABLE Season (
    id INTEGER,
    player_id INTEGER,
    team_id INTEGER,
    year CHAR(4),
    PRIMARY KEY (player_id, team_id, year) 
);
''')
conn.commit()
```

### Create table of NBA teams

```python
html = urlopen('https://www.basketball-reference.com/teams/')
html_read = html.read()
bsObj = BeautifulSoup(html_read)

for child in bsObj.find('tbody').children:
    try:
        team_url = child.find('a').get('href')
        team_code = re.split('/', team_url)[2] # extract 3-digit team code from URL
        team_name = child.find('th').get_text() # get full team name
    except:
        continue
        
    # fix abbreviation for teams with with recently changed abbreviations
    if team_code == 'NJN':
        team_code = 'BRK'
    elif team_code == 'NOH':
        team_code = 'NOP'
    elif team_code == 'CHA':
        team_code = 'CHO'
        
    # add data to NBA_Fantasy_db
    cur.execute('''INSERT INTO TEAM (name, code)
                   VALUES (?, ?)''', (team_name, team_code))
    conn.commit()
```

### Create table of player salaries for each year

```python
cur.execute('''SELECT name_ascii FROM Player''')
names_ascii = cur.fetchall()
names_ascii = [name[0] for name in names_ascii]
```

```python
def get_salaries_from_table(salary_table):
    for row in salary_table.children:            
        data = list(row.children) # data contains four values (id, name, team, salary)
        for i in range(len(data)):
            data[i] = data[i].get_text() # remove html code
            
        if data[0] == 'RK': # if this is a header row, ignore
            continue
            
        data[1] = re.split(',', data[1])[0] # get rid of player position at the end of the string
        data[3] = re.split('\$', data[3])[-1] # get rid of $ sign in salary
        data[3] = int(''.join(re.split(',', data[3]))) # get rid of commas in salary and convert to int
        
        player_name = data[1]
        team_name = data[2]
        amount = data[3]
        
        # fix inconsistencies in names across diferent websites
        if team_name == 'LA Clippers':
            team_name = 'Los Angeles Clippers'
        if player_name == 'Otto Porter Jr.':
            player_name = 'Otto Porter'
        if player_name == 'JR Smith':
            player_name = 'J.R. Smith'
            
        # add column for ascii version of name to Player
        
        for name in names_ascii:
            if re.match(name, player_name):
                cur.execute('''UPDATE Player SET name_ascii = ? WHERE name_ascii == ?''', (player_name, name))
               
        try:
            cur.execute('''SELECT id FROM Player WHERE name_ascii == ?''', (player_name,))
            player_id = cur.fetchone()[0]
        
            cur.execute('''INSERT OR IGNORE INTO Salary (player_id, season_id, amount)
                           VALUES (?, ?, ?)''', (player_id, season_id, amount))
        except:
            if player_name not in failed:
                failed.append(player_name)
    conn.commit()
```

# Data Cleaning

```python
df_clean = df_stats.copy()

# convert columns with numbers to numeric type
numeric_columns = ['FG', 'FGA', '3P', '3PA', 'FT', 'FTA', 'ORB','DRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', '+/-',
                   'home', 'team_Pace', 'team_eFG%', 'team_TOV%', 'team_ORB%', 'team_FT/FG', 'team_ORtg','opp_Pace', 
                   'opp_eFG%', 'opp_TOV%', 'opp_ORB%', 'opp_FT/FG', 'opp_ORtg', 'TS%', 'eFG%', '3PAr', 'FTr', 'ORB%', 
                   'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%', 'ORtg', 'DRtg', 'BPM',
                   'team_inactive_salary', 'opp_inactive_salary'
                  ]
for column in numeric_columns:
    df_clean[column] = pd.to_numeric(df_clean[column])

# convert min from string to numeric
def min_to_numeric(value):
    minute = int(re.split(':', value)[0])
    second = int(re.split(':', value)[1])
    return minute + round(second / 60, 2)
df_clean['min'] = df_clean['min'].apply(lambda x: min_to_numeric(x))

# convert date to datetime
df_clean['date'] = pd.to_datetime(df_clean['date'])

# only consider players who played for at least 25 min
# I am not interested in making predictions for low-scoring players
df_clean = df_clean[df_clean['min'] >= 25]

df_clean.head()
```

# Create dataset with following data:
- player stats averaged over season up until game of interest
- player stats averaged over the last week
- team/opp stats averaged over season up until game of interest
- team/opp stats averaged over the last week


```python
df_final = df_clean[['player_name', 'team_name', 'opp_name', 'home', 'date', 'team_inactive_salary', 
                     'opp_inactive_salary', 'PTS']].copy()

# only consider games starting in January so that we have at least a week of games already played
df_final = df_final[df_final['date'] >= pd.to_datetime('2021-01-01')]

# add player stats averaged over the last three games
def player_week(row):
    player = row['player_name']
    team = row['team_name']
    date = row['date']
    df_player = df_clean[df_clean['player_name'] == player]
    df_player_week = df_player[(df_player['date'] < date) & (df_player['date'] >= date - pd.to_timedelta(1, unit='W'))]
    if len(df_player_week) == 0:
        col = df_player.drop(columns=['player_name','date','team_name','opp_name', 'team_abbrev', 'opp_abbrev']).columns
        AVGs = pd.Series(np.zeros(len(df_stats.columns) - 6), index=col)
    else:
        AVGs = df_player_week.drop(columns=['player_name','date','team_name','opp_name', 'team_abbrev', 'opp_abbrev']).mean()
    return AVGs

def player_season(row):
    player = row['player_name']
    team = row['team_name']
    date = row['date']
    df_player = df_clean[df_clean['player_name'] == player]
    df_player_season = df_player[df_player['date'] < date]
    if len(df_player_season) == 0:
        col = df_player.drop(columns=['player_name','date','team_name','opp_name', 'team_abbrev', 'opp_abbrev']).columns
        AVGs = pd.Series(np.zeros(len(df_stats.columns) - 6), index=col)
    else:
        AVGs = df_player_season.drop(columns=['player_name','date','team_name','opp_name', 'team_abbrev', 'opp_abbrev']).mean()
    return AVGs

def opp_week(row):
    player = row['player_name']
    opp = row['opp_name']
    date = row['date']
    df_opp = df_clean[df_clean['team_name'] == opp]
    df_opp = df_opp[['date','team_Pace', 'team_eFG%', 'team_TOV%', 'team_ORB%', 'team_FT/FG', 'team_ORtg', 
                     'opp_Pace', 'opp_eFG%', 'opp_TOV%', 'opp_ORB%', 'opp_FT/FG', 'opp_ORtg']]
    df_opp_week = df_opp[(df_opp['date'] < date) & (df_opp['date'] >= date - pd.to_timedelta(1, unit='W'))]
    if len(df_opp_week) == 0:
        col = df_opp_week.drop(columns=['date']).columns
        AVGs = pd.Series(np.zeros(len(df_opp.columns) - 1), index=col)
    else:
        df_opp_week = df_opp_week.groupby(by='date').agg(np.mean)
        AVGs = df_opp_week.mean()
    return AVGs
        
def opp_season(row):
    player = row['player_name']
    opp = row['opp_name']
    date = row['date']
    df_opp = df_clean[df_clean['team_name'] == opp]
    df_opp = df_opp[['date','team_Pace', 'team_eFG%', 'team_TOV%', 'team_ORB%', 'team_FT/FG', 'team_ORtg', 
                     'opp_Pace', 'opp_eFG%', 'opp_TOV%', 'opp_ORB%', 'opp_FT/FG', 'opp_ORtg']]
    df_opp_season = df_opp[df_opp['date'] < date]
    if len(df_opp_season) == 0:
        col = df_opp_season.drop(columns=['date']).columns
        AVGs = pd.Series(np.zeros(len(df_opp.columns) - 1), index=col)
    else:
        df_opp_season = df_opp_season.groupby(by='date').agg(np.mean)
        AVGs = df_opp_season.mean()
    return AVGs
    
df_player_week = df_final.apply(lambda x: player_week(x), axis=1)
df_player_season = df_final.apply(lambda x: player_season(x), axis=1)
df_opp_week = df_final.apply(lambda x: opp_week(x), axis=1)
df_opp_season = df_final.apply(lambda x: opp_season(x), axis=1)
```

### Get rid of rows with lots of 0s
- player was likely coming off injury, which might skew predictions unnecessarily
- could also represent game right after all-star break where game had not been played in over a week

```python
def number_zeroes(row):
    values = row.values
    flag = 0
    for value in values:
        if value == 0:
            flag += 1
    if flag > 10:
        return 1
    else:
        return 0

mask = df_player_week.apply(lambda x: number_zeroes(x), axis=1)
mask = mask.values
mask = mask == 0

df_player_week = df_player_week.iloc[mask]
df_player_season = df_player_season.iloc[mask]
df_opp_week = df_opp_week.iloc[mask]
df_opp_season = df_opp_season.iloc[mask]
```

```python
df_player_week = df_player_week - df_player_season
df_opp_week = df_opp_week - df_opp_season
```

```python
df_final = pd.merge(df_final, df_player_week, how='inner', left_index=True, right_index=True, suffixes=('','_player_week_AVG'))
df_final = pd.merge(df_final, df_player_season, how='inner', left_index=True, right_index=True, suffixes=('','_player_season_AVG'))
df_final = pd.merge(df_final, df_opp_week, how='inner', left_index=True, right_index=True, suffixes=('','_opp_week_AVG'))
df_final = pd.merge(df_final, df_opp_season, how='inner', left_index=True, right_index=True, suffixes=('','_opp_season_AVG'))
```

# Create data to predict during NBA playoffs
- this function will take information about a player for a particular game and output a prediction using the second model to compare to a over/under generated by a betting site

```python
team_salaries = df_players.groupby(by='team_name')['salary'].agg(np.sum)
opp_salaries = team_salaries['Phoenix Suns']
opp_salaries
```

```python
def predict_player_points(player, team, opp, home, date, team_injuries, opp_injuries):
    X_predict = pd.DataFrame(np.array([date, team, home, opp]).reshape(1,-1), columns = ['date', 'team', 'home', 'opp'])
    
    # determine salaries of injured players
    team_salary = []
    opp_salary = []
    for injured_player in team_injuries:
        team_salary.append(df_players[df_players['player_name'] == injured_player]['salary'].values[0])
    for injured_player in opp_injuries:
        opp_salary.append(df_players[df_players['player_name'] == injured_player]['salary'].values[0])
    
    team_salaries = df_players.groupby(by='team_name')['salary'].agg(np.sum)
    team_salary_sum = team_salaries[team]
    opp_salary_sum = team_salaries[opp]
    team1_inactive_salary = np.sum(team_salary)/team_salary_sum
    team2_inactive_salary = np.sum(opp_salary)/opp_salary_sum
    df_inactive_salary = pd.DataFrame(np.array([team1_inactive_salary, team2_inactive_salary]).reshape(1,-1),
                                      columns=['team_inactive_salary', 'opp_inactive_salary'])
    
    # add player stats averaged over the last week
    df_player = df_clean[df_clean['player_name'] == player]
    df_player_week = df_player[(df_player['date'] < date) & (df_player['date'] >= date - pd.to_timedelta(1, unit='W'))]
    df_player_week = df_player_week.drop(columns=['player_name','date','team_name','opp_name', 'team_abbrev', 'opp_abbrev']).mean()
    df_player_week = pd.DataFrame(np.array(df_player_week).reshape(1,-1), columns = df_player_week.index)
    
    # add player stats averaged over the season
    df_player_season = df_player[df_player['date'] < date]
    df_player_season = df_player_season.drop(columns=['player_name','date','team_name','opp_name', 'team_abbrev', 'opp_abbrev']).mean()
    df_player_season = pd.DataFrame(np.array(df_player_season).reshape(1,-1), columns = df_player_season.index)
    
    # add opponent stats averaged over the last week
    df_opp = df_clean[df_clean['team_name'] == opp]
    df_opp = df_opp[['date','team_Pace', 'team_eFG%', 'team_TOV%', 'team_ORB%', 'team_FT/FG', 'team_ORtg', 
                     'opp_Pace', 'opp_eFG%', 'opp_TOV%', 'opp_ORB%', 'opp_FT/FG', 'opp_ORtg']]
    df_opp_week = df_opp[(df_opp['date'] < date) & (df_opp['date'] >= date - pd.to_timedelta(1, unit='W'))]
    df_opp_week = df_opp_week.groupby(by='date').agg(np.mean)
    df_opp_week = df_opp_week.mean()
    df_opp_week = pd.DataFrame(np.array(df_opp_week).reshape(1,-1), columns = df_opp_week.index)
        
    # add opponent stats averaged over the season
    df_opp_season = df_opp[df_opp['date'] < date]
    df_opp_season = df_opp_season.groupby(by='date').agg(np.mean)
    df_opp_season = df_opp_season.mean()
    df_opp_season = pd.DataFrame(np.array(df_opp_season).reshape(1,-1), columns = df_opp_season.index)
    
    # subtract week from season
    df_player_week = df_player_week - df_player_season
    df_opp_week = df_opp_week - df_opp_season

    # merge all three df's
    X_predict = pd.merge(X_predict, df_inactive_salary, how='inner', left_index=True, right_index=True)
    X_predict = pd.merge(X_predict, df_player_week, how='inner', left_index=True, right_index=True, suffixes=('','_player_week_AVG'))
    X_predict = pd.merge(X_predict, df_player_season, how='inner', left_index=True, right_index=True, suffixes=('','_player_season_AVG'))
    X_predict = pd.merge(X_predict, df_opp_week, how='inner', left_index=True, right_index=True, suffixes=('','_opp_week_AVG'))
    X_predict = pd.merge(X_predict, df_opp_season, how='inner', left_index=True, right_index=True, suffixes=('','_opp_season_AVG'))
    # drop team and opp
    X_predict.drop(columns=['date','team','opp'], axis=1, inplace=True)

    for col in X_predict.columns:
        if X_predict[col].isnull().sum() == len(X_predict):
            X_predict.drop(columns=[col], axis=1, inplace=True)
            
    X_predict = scale.transform(X_predict)
    y_pred = np.round(vote.predict(X_predict), 1)
    return y_pred[0]
    
```

# Statistical Analysis

```python
from scipy.stats import ttest_ind, bayes_mvs, bartlett

stat, p = ttest_ind(data['Diff'], data['O/U Diff'])

print('The p-value for the T-test between the model and O/U differences from actual points scored is: {}'.format(p))
if p < 0.05:
    print('There IS a significant difference between the model and O/U')
else:
    print('There is NOT a significant difference between the model and O/U')
    
Model_stats = bayes_mvs(data['Diff'], alpha=0.95)
OU_stats = bayes_mvs(data['O/U Diff'], alpha=0.95)

print('')
print('The 95% CI for the model prediction difference from actual is {}'.format(Model_stats[0][1]))
print('The 95% CI for the O/U from actual is {}'.format(OU_stats[0][1]))
```

```python
stat, p = bartlett(data['Diff'], data['O/U Diff'])
print('The p-value for Bartletts test for equal variances between the model and O/U')
print('differences from actual points scored is: {}'.format(p))

if p < 0.05:
    print('There IS a significant difference between the variation of the model and O/U')
else:
    print('There is NOT a significant difference between the variation of the model and O/U')
```

Because there is no significant difference in the mean and standard deviation of the model and O/U predictions on points scored, the model I developed is comparable to the O/U lines set by monkeyknifefight.com.
